```python
"""
This script performs Exploratory Data Analysis (EDA) on a merged dataset
containing allele frequencies and UV radiation data by population.

Instructions:
-------------
1. Make sure the merged CSV file exists and that the path is correct for 
your working directory.

2. Run this script using a Python interpreter (or inside Jupyter as a code cell, if desired).

"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the merged dataset
file_path = "C:/Users/maeso/OneDrive/Documents/merged_allele_uv_data.csv"
df = pd.read_csv(file_path)

# Preview the data
print("First 5 rows of the dataset:")
print(df.head())

# Check for missing values
print("\n Missing values per column:")
print(df.isnull().sum())

# Summary statistics
print("\n Summary statistics:")
print(df.describe())

# Distribution of UV Index
plt.figure(figsize=(8, 5))
sns.histplot(df['Annual_UV_Index'].dropna(), bins=10, kde=True, color="orange")
plt.title("Distribution of Annual UV Index by Population")
plt.xlabel("Annual UV Index")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Scatter plot: AA_Freq vs. Annual UV Index
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Annual_UV_Index', y='AA_Freq', hue='Population_Name', palette='tab10')
plt.title("AA_Freq vs. Annual UV Index")
plt.xlabel("Annual UV Index")
plt.ylabel("AA Allele Frequency")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Correlation matrix (for numeric columns)
print("\n Correlation matrix:")
print(df[['AA_Freq', 'A_Freq', 'Annual_UV_Index']].corr())

# Heatmap of correlation
plt.figure(figsize=(6, 4))
sns.heatmap(df[['AA_Freq', 'A_Freq', 'Annual_UV_Index']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
